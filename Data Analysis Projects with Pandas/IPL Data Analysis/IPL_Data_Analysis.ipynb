## 1. To find the matches played in each season

```python
match_per_season = match_data.groupby(['Season'])['id'].count().reset_index().rename(columns={'id':'matches'})
match_per_season 
```

```python
match_data.head(2)
```

```python
sns.countplot(match_data['Season'])     # to plot the graph for Total number of matches played in each season
```

```python
sns.countplot(match_data['Season'])   
plt.xticks(rotation=45, fontsize=10)
plt.yticks(rotation=45, fontsize=10)
plt.xlabel('Seasons', fontsize=10)
plt.xlabel('Number of Matches', fontsize=10)
plt.title('Total number of matches played in each season',fontsize=15, fontweight='bold')
```

```python
# here we are creating a new data by merging 'match_data' and 'ball_data'
season_data=match_data[['id','Season']].merge(ball_data, left_on='id', right_on='id', how = 'left').drop('id', axis=1)
season_data.head()
```

## 2. here we are combining the total number of runs scored in each season and plotting a graph


```python
season=season_data.groupby(['Season'])['total_runs'].sum().reset_index()
p=season.set_index('Season')
ax = plt.axes()
ax.set(facecolor='black')
sns.lineplot(data=p, color='r*-')
plt.title('Total number of runs scored in each season ', fontsize=12,fontweight='bold')
plt.show()
```

## 3. To find the average runs scored per match in each season

```python
runs_per_match= match_per_season.concat([match_per_season,season.iloc[:,-1]],axis=1)

runs_per_match
```

```python
runs_per_match['Runs scored per match']=runs_per_match['total_runs']/runs_per_match['matches']
runs_per_match
```

```python
runs_per_match.set_index('Season', inplace=True)
runs_per_match
```
